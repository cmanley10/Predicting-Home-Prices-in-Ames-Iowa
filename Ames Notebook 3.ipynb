```python
#function to turn garage year built into categories based on the following years, with numerical values
def yeartocat(year):
    if year != year:
        return 0
    elif year < 1961:
        return 1
    elif year < 1980:
        return 2
    elif year < 2002:
        return 3
    else:
        return 4
```

```python

train['Garage Yr Blt'] = train['Garage Yr Blt'].map(yeartocat)
```

```python
def year_built_cat(year):
    if year != year:
        return 0
    elif year < 1953:
        return 1
    elif year < 1974:
        return 2
    elif year < 2001:
        return 3
    else:
        return 4
```
